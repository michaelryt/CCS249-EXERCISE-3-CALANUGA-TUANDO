```python
# Import Necessary Libraries
import wikipedia
import re
import math
import nltk
from nltk import bigrams, trigrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Download necessary NLTK data
nltk.download('punkt')
```

1.	 (10 points) Use the Wikipedia python module and access any topic, as you will use that as your corpus, with a word limit of 1000 words.

```python
# Access Wikipedia content (Limit to 1000 words)
topic = "EDSA People Power"
page = wikipedia.page(topic)
text = ' '.join(page.content.split()[:1000])  # Limit to 1000 words

# Clean text: Remove punctuation & convert to lowercase
text = re.sub(r'[^\w\s]', '', text).lower()

# Tokenization
tokens = word_tokenize(text)

# Print the cleaned content of the page
print(text)
```

2.	(60 points) Train 2 models: a Bigram and Trigram Language Model, use the shared code as reference for bigram modeling, and update it to support trigrams.

```python
# Bigram Probability Model with Laplace Smoothing
def bigram_probabilities(tokens, alpha=1):
    bigram_counts = Counter(bigrams(tokens))
    unigram_counts = Counter(tokens)
    vocab_size = len(unigram_counts)  # Vocabulary size for smoothing

    bigram_probs = {
        bigram: (count + alpha) / (unigram_counts[bigram[0]] + alpha * vocab_size)
        for bigram, count in bigram_counts.items()
    }
    return bigram_probs
wiki_bigram = bigram_probabilities(text)
print(wiki_bigram)
```

```python

# Trigram Probability Model with Laplace Smoothing
def trigram_probabilities(tokens, alpha=1):
    trigram_counts = Counter(trigrams(tokens))
    bigram_counts = Counter(list(bigrams(tokens)))
    vocab_size = len(bigram_counts)  # Vocabulary size for smoothing

    trigram_probs = {
        trigram: (count + alpha) / (bigram_counts.get((trigram[0], trigram[1]), 0) + alpha * vocab_size)
        for trigram, count in trigram_counts.items()
    }
    return trigram_probs
wiki_trigram = trigram_probabilities(text)
print(wiki_trigram)
```

```python
# Train Models
wiki_bigram = bigram_probabilities(tokens)
wiki_trigram = trigram_probabilities(tokens)
```

3.	(30 points) Using a test sentence “The quick brown fox jumps over the lazy dog near the bank of the river.” OR generate your own test sentence, create a function that will determine the perplexity score for each trained model.

```python
# Bigram Perplexity Calculation
def compute_bigram_perplexity(bigram_probs, test_sentence, vocab_size, alpha=1):
    test_tokens = word_tokenize(test_sentence.lower())
    test_bigrams = list(bigrams(test_tokens))
    
    N = len(test_bigrams)
    log_prob_sum = 0

    for bigram in test_bigrams:
        prob = bigram_probs.get(bigram, alpha / (vocab_size + alpha * vocab_size))  # Laplace smoothing
        log_prob_sum += math.log2(prob)
    
    return math.pow(2, -log_prob_sum / N) if N > 0 else float('inf')
```

```python
# Trigram Perplexity Calculation
def compute_trigram_perplexity(trigram_probs, test_sentence, vocab_size, alpha=1):
    test_tokens = word_tokenize(test_sentence.lower())
    test_trigrams = list(trigrams(test_tokens))
    
    N = len(test_trigrams)
    log_prob_sum = 0

    for trigram in test_trigrams:
        prob = trigram_probs.get(trigram, alpha / (vocab_size + alpha * vocab_size))  # Laplace smoothing
        log_prob_sum += math.log2(prob)
    
    return math.pow(2, -log_prob_sum / N) if N > 0 else float('inf')
```

```python
# Test Sentence
test_sentence = "The EDSA People Power Revolution was a peaceful uprising in 1986 that led to the restoration of democracy in the Philippines"
```

```python
# Compute Perplexity Scores
bigram_perplexity = compute_bigram_perplexity(wiki_bigram, test_sentence, len(tokens))
trigram_perplexity = compute_trigram_perplexity(wiki_trigram, test_sentence, len(tokens))
```

```python
# Print results
print("\n--- Perplexity Scores ---")
print("Bigram Model Perplexity -> Score:", bigram_perplexity)
print("Trigram Model Perplexity -> Score:", trigram_perplexity)
```
